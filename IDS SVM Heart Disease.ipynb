```python
# Initializing the data-set

import pandas as pd

hd = pd.read_csv('heart.csv')

hd.head()
```

```python
#List number of values

hd['target'].value_counts()
```

```python
# Here we split the data-set into training and test parts

from sklearn.model_selection import train_test_split
X = hd.drop('target', axis=1)
y = hd['target']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=None)
```

```python
# Test dataset value counts

y_test.value_counts()
```

```python
# In this part, we initialize SVC

from sklearn.svm import SVC
svm_model=SVC()
```

```python
# This is the part where the SVC model is trained

svm_model.fit(X_train,y_train)
```

```python
predictions = svm_model.predict(X_test)
```

```python
# For statistics report we will use precision, recall and accuracy from sklearn.metrics

from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
# Report before SVM

print(accuracy_score(y_test, predictions))
print(recall_score(y_test, predictions))
print(precision_score(y_test, predictions))
```

```python
# Tweaking the parameters

param_grid = {'C':[0.01,0.1,1,10,100,1000,10000,100000],
              'gamma':[1,0.1,0.01,0.001,0.0001,0.00001,0.000001]}

from sklearn.model_selection import GridSearchCV

# We will user GridSearchCV in order to 

grid = GridSearchCV(SVC(),param_grid,verbose=2)
```

```python
grid.fit(X_train,y_train)
```

```python
grid_predictions = grid.predict(X_test)

# Report after SVM algorithm
print("Accuracy")
print(accuracy_score(y_test, grid_predictions))
print ("Recall")
print(recall_score(y_test, grid_predictions))
print("Precision")
print(precision_score(y_test, grid_predictions))
```
